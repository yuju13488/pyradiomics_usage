```python
import os
from dcmrtstruct2nii import dcmrtstruct2nii, list_rt_structs
import numpy as np
import nibabel as nib
import nrrd
import dicom2nifti
from pydicom import dcmread
import time

nii_rs = "./data/nii_rs/"
nrrd_rs = "./data/nrrd_rs/"
nii_ct = "./data/nii_ct/"
nrrd_ct = "./data/nrrd_ct/"
```

```python
# RS file of meta
path_meta = './data/Meta/'
rs_meta_list = []
for dirPath, dirNames, fileNames in os.walk(path_meta):
    for f in fileNames:
        if f[:2] == "RS":
            rs_meta_list.append(os.path.join(dirPath, f))

# make nii file and nrrd file
for i in range(len(rs_meta_list)):
    print("第 {} 位".format(i + 1))
    rs_path = rs_meta_list[i]
    rs_index = rs_path.find("RS")
    ct_path = rs_path[:rs_index]
    ds = dcmread(rs_path)
    PatientID = ds["PatientID"].value
    print(PatientID)
    nii_rs_path = nii_rs + "meta/" + PatientID + "/" # path of file saving
    if not os.path.exists(nii_rs_path):
        os.makedirs(nii_rs_path)
    
    # probability name of lymph nodes
    ROI_list = list_rt_structs(rs_path)
    if "GTV-LN" in ROI_list:
        ROI_need = "GTV-LN"
    elif "GTV-N" in ROI_list:
        ROI_need = "GTV-N"
    elif "_LN" in ROI_list:
        ROI_need = "LN"
    elif "GTVN" in ROI_list:
        ROI_need = "GTVN"
    elif "GTVn" in ROI_list:
        ROI_need = "GTVn"
    else:
        ROI_need = "No lymph nodes!"
    print(ROI_need)
        
    try:
        # specific ROI to nii file, image.nii.gz aslo from all CT dicom file
        dcmrtstruct2nii(rs_path, ct_path, nii_rs_path)
    except:
        print("No " + ROI_need + "!")
    # RS nii file to nrrd file
    files_roi = os.listdir(nii_rs_path)
    for f in files_roi:
        example_filename = nii_rs_path + f
        img = nib.load(example_filename)
        array = np.array(img.dataobj)
        nrrd_path_to = nrrd_rs + "meta/" + PatientID + "/" # path of file saving
        if not os.path.exists(nrrd_path_to):
            os.makedirs(nrrd_path_to)
        nrrd.write(nrrd_path_to + f.replace("nii.gz", "nrrd"), array)
    # CT dicom file to nii file
    nii_ct_path = nii_ct + "meta/" + PatientID + "/" # path of file saving
    if not os.path.exists(nii_ct_path):
        os.makedirs(nii_ct_path)
    dicom2nifti.convert_directory(ct_path, nii_ct_path)
    # CT nii file to nrrd file
    files = os.listdir(nii_ct_path)
    for f in files:
        if f.find("nii") != -1:
            nii_file = f
    example_filename = nii_ct_path + nii_file
    img = nib.load(example_filename)
    array = np.array(img.dataobj)
    nrrd_path_to = nrrd_ct + "meta/" + PatientID + "/" # path of file saving
    
    if not os.path.exists(nrrd_path_to):
        os.makedirs(nrrd_path_to)
    nrrd.write(nrrd_path_to + f.replace("nii.gz", "nrrd"), array)
    print("--------------------------------------")
```
