```python
import pandas
import numpy as np
import os
from scipy import stats
from collections import Counter
```

```python
feature_meta_path = "./data/txt/meta/"
files_meta = os.listdir(feature_meta_path)

feature_nonmeta_path = "./data/txt/nonmeta/"
files_nonmeta = os.listdir(feature_nonmeta_path)

p_dict = {}
c = 0
while True:
    try:
        c += 1
        feature_name_meta = []
        feature_name_nonmeta = []
        feature_value_meta = []
        feature_value_nonmeta = []
        # get name of feature, value of feature
        for file_m in files_meta:
            with open(feature_meta_path + file_m, 'r', encoding = 'utf-8') as f:
                features_m_all = f.read()
            # split "name of feature : value of feature"
            feature_m_all = features_m_all.split("\n")
            voi_m = 1
            feature_m = feature_m_all[c].split(":")
            # name of feature
            feature_name_m = feature_m[0]
            feature_name_meta.append(feature_name_m)
            try:
                # value of feature
                feature_value_m = float(feature_m[1]) / voi_m
                feature_value_meta.append(feature_value_m)
            # if value of feature not numerical
            except:
                feature_value_m = 0
                continue
        # check
        if list(Counter(feature_name_meta).values())[0] != len(files_meta):
            print("error!")

        for file_n in files_nonmeta:
            with open(feature_nonmeta_path + file_n, 'r', encoding = 'utf-8') as f:
                features_n_all = f.read()
            feature_n_all = features_n_all.split("\n")
            voi_n = 1
            feature_n = feature_n_all[c].split(":")
            feature_name_n = feature_n[0]
            feature_name_nonmeta.append(feature_name_n)
            try: 
                feature_value_n = float(feature_n[1]) / voi_n
                feature_value_nonmeta.append(feature_value_n)
            except:
                feature_value_n = 0
                continue
        if list(Counter(feature_name_nonmeta).values())[0] != len(files_nonmeta):
            print("error!")
        # get p-value by T-test
        if len(feature_value_meta) > 0 and len(feature_value_nonmeta) > 0:
            t, p = stats.ttest_ind(feature_value_meta, feature_value_nonmeta)
            print(feature_name_nonmeta[0] + " : ")
            print("length of feature_value_meta:", len(feature_value_meta))
            print("length of feature_value_nonmeta:", len(feature_value_nonmeta))
            print("max:", max(feature_value_meta+feature_value_nonmeta))
            print("min:", min(feature_value_meta+feature_value_nonmeta))
            print("p-value:", p)
            if p <= 0.1:
                p_dict[feature_name_nonmeta[0]] = p
            print('------------------------------------------------------')
            print()
    except IndexError:
        break
```
